```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("../data/cleaned_student_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# STEP 4: Understand the Target Variable (Passed)
# Class Distribution
sns.countplot(x="Passed", data=df)
plt.title("Distribution of Pass vs Fail")
plt.show()
```

```python
# STEP 5: Univariate Analysis (One Feature at a Time)
# Study Hours Distribution
sns.histplot(df["Study Hours per Week"], kde=True)
plt.title("Distribution of Study Hours per Week")
plt.show()
```

```python
# Attendance Rate Distribution
sns.histplot(df["Attendance Rate"], kde=True)
plt.title("Attendance Rate Distribution")
plt.show()
```

```python
# Study Hours vs Passed
sns.boxplot(x="Passed", y="Study Hours per Week", data=df)
plt.title("Study Hours vs Pass/Fail")
plt.show()
```

```python
# STEP 7: Categorical Feature Impact (Encoded Columns)

# Since you used one-hot encoding, columns look like:

# Participation_in_Extracurricular_Activities_Yes

# Parent_Education_Level_Master (example)

sns.boxplot(
    x="Passed",
    y="Participation in Extracurricular Activities_Yes",
    data=df
)
plt.title("Extracurricular Activities vs Pass/Fail")
plt.show()
```

```python
# STEP 8: Correlation Analysis (VERY IMPORTANT)
#  Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Key Insights from EDA

- Study hours show a mild positive influence on passing.
- Attendance rate does not strongly differentiate pass and fail cases.
- Previous grades have some predictive value.
- No single feature dominates prediction.
